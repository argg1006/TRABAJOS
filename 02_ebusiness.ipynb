## Parte 1: Análisis de datos sintéticos

En esta parte vamos a analizar un dataset de transacciones y determinar si la estrategia de M-Commerce (App Móvil) es más rentable que el E-Commerce tradicional (Web Desktop).

Vamos a generar una base de datos aleatoria, que simula la salida de un TPS.

```python
import pandas as pd
import numpy as np

# Generar datos simulados
data = {
    'ID_Transaccion': range(1001, 1051),
    'Canal': np.random.choice(['Web Desktop', 'App Móvil'], 50, p=[0.4, 0.6]),
    'Metodo_Pago': np.random.choice(['Tarjeta Crédito', 'PayPal', 'Transferencia'], 50),
    'Total_Venta': np.random.randint(20, 500, 50), # Ventas entre $20 y $500
    'Categoria': np.random.choice(['Electrónica', 'Moda', 'Hogar'], 50)
}

df = pd.DataFrame(data)

df
```

### ¿Qué preguntas deberíamos plantearnos para obtener las respuestas que buscamos?

Buscamos saber cuál de las dos estrategias (M-Commerce o E-Commerce) es la más rentable.

Vamos a crear 4 KPIs para un MIS que nos permitan responder esta pregunta.

```python
# KPI 1: Ingreso Total por Canal
kpi1 = df.groupby('Canal')['Total_Venta'].sum()
print("Ingreso Total por Canal ")
print(kpi1)
print()

# KPI 2: Ticket Promedio por Canal
kpi2 = df.groupby('Canal')['Total_Venta'].mean().round(2)
print("Ticket Promedio por Canal")
print(kpi2)
print()

# KPI 3: Número de Transacciones por Canal
kpi3 = df.groupby('Canal')['ID_Transaccion'].count()
print("=== KPI 3: Número de Transacciones por Canal ===")
print(kpi3)
print()

# KPI 4: Ingreso por Categoría y Canal
kpi4 = df.groupby(['Canal', 'Categoria'])['Total_Venta'].sum()

print("Ingreso por Categoría y Canal")
print(kpi4)
```

**Conclusiones**


*   En base a los resultados dados, el más rentable es M-Commerce que el E-Commerce, por su comparación con los KPIS, se refleja que tiene mas ingresos por canal, a su vez el número de transacciones, e igual puede haber leves difencias en ingreso por categoría, pero sigue este teniendo más ingresos.

### 1. KPIs de Gestión de Logística y Operaciones (Shipment & Warehouse)

Permiten medir la eficiencia en la distribución de los productos electrónicos.

1.1. **Costo Promedio de Envío por Modo de Transporte (Mode of Shipment):** Ayuda a evaluar qué canales (Barco, Avión, Carretera) representan un mayor impacto financiero en relación con el peso del producto.

1.2. **Volumen de Ocupación por Bloque de Bodega (Warehouse Block):** Porcentaje o cantidad total de productos/peso administrados en cada bloque (A, B, C, D, E) para entender la carga operativa de la infraestructura.

1.3. **Peso Promedio del Producto por Modo de Envío:** Para analizar si se está optimizando la logística (ej. si los productos más pesados se están enviando por barco o si se está gastando de más enviándolos por avión).

```python
import pandas as pd
data = pd.read_csv('Train.csv')

#1.1 Costo Promedio de Envio por Modo de Transporte
Promedio=data.groupby('Mode_of_Shipment')['Cost_of_the_Product'].mean().round(2)
print("Costo Promedio de Envio por Modo de Transporte ")
print(Promedio)

#1.2 Volumen de Ocupación por Bloque de Bodega
Volumen=data.groupby('Warehouse_block')['ID'].count()
print("Volumen de Ocupación por Bloque de Bodega")
print(Volumen)

#1.3 Peso Promedio del Producto por Modo de Envío
Peso=data.groupby('Mode_of_Shipment')['Weight_in_gms'].mean().round(2)
print("Peso Promedio del Producto por Modo de Envío")
print(Peso)
```

**CONCLUSIONES:**

1.1 -Los tres modos tienen costos casi idénticos lo que sugiere que el precio del producto no influye en la elección del medio de envío.

1.2 -El bloque F tiene el doble de productos que todos los demás indicando que F está sobrecargado y podría ser un cuello de botella.

1.3 -El peso es casi igual en los tres modos, esto es una ineficiencia logística, lo ideal sería que el barco concentre los productos más pesados, por su capacidad.

### 2. KPIs de Gestión Comercial y Financiera (Sales & Revenue)

Enfocados en el rendimiento del dinero, los descuentos y la estrategia de precios.

2.1. **Ingreso Total Bruto y Costo Promedio del Producto:** Suma y promedio de la variable `Cost of the product` para entender el ticket promedio de la tienda de electrónica.

2.2. **Porcentaje de Descuento Promedio Otorgado:** Evaluar el impacto de la estrategia de promociones (`Discount offered`) y si los descuentos altos realmente correlacionan con clientes de mayor prioridad o compras previas.

2.3. **Margen Teórico Relativo por Importancia del Producto:** Cruzar `Product importance` (Low, Medium, High) con el costo para ver si los productos catalogados como "alta importancia" son también los más costosos o los que reciben más descuentos.

```python
import pandas as pd
data = pd.read_csv('Train.csv')
#2.1 Ingreso Total Bruto y Costo Promedio
total = data['Cost_of_the_Product'].sum()
promedio = data['Cost_of_the_Product'].mean().round(2)
print("Ingreso Total Bruto y Costo Promedio")
print(f"Total bruto: ${total:,}")
print(f"Costo promedio: ${promedio}")

#2.2: Descuento Promedio según Compras Previas
Descuento = data.groupby('Prior_purchases')['Discount_offered'].mean().round(2)
print("Descuento Promedio según Compras Previas")
print(Descuento)

#KPI 2.3: Costo y Descuento por Importancia del Producto
Importancia = data.groupby('Product_importance')[['Cost_of_the_Product', 'Discount_offered']].mean().round(2)
print("Costo y Descuento por Importancia del Producto")
print(Importancia)
```

**CONCLUSIONES:**

2.1 -Tanto el ingreso bruto total como un ticket promedio, es un número estable que sirve como línea base para medir el desempeño.

2.2 -El patrón es irregular siendo que no hay una tendencia clara, la estrategia de descuentos no está bien segmentada siendo que no premia consistentemente la lealtad.

2.3 -Los productos de alta importancia  son los más baratos y los de baja importancia son los más costosos. Eso indica que la clasificación de importancia no está alineada con el precio.

### 3. KPIs de Experiencia y Fidelización del Cliente (Customer Experience - CX)

Diseñados para evaluar la satisfacción y la carga de soporte post-venta.

3.1. **Índice de Satisfacción del Cliente (CSAT) Promedio:** El promedio de la variable `Customer rating` (escala 1 a 5). Se puede segmentar por género o por bloque de bodega para ver si hay anomalías.

3.2. **Tasa de Retención / Clientes Recurrentes:** Analizar la distribución de `Prior purchases` para determinar qué porcentaje de la base de datos son clientes nuevos (0-1 compras previas) vs. clientes fidelizados (2 o más compras).

3.3. **Carga de Soporte por Cliente (Customer Care Calls):** Promedio de llamadas de consulta realizadas. Permite analizar si a mayor costo del producto o menores calificaciones, aumentan las llamadas de reclamo.

```python
import pandas as pd
data = pd.read_csv('Train.csv')
#3.1: CSAT Promedio por Género y Bloque de Bodega
Genero = data.groupby('Gender')['Customer_rating'].mean().round(2)
Bodega = data.groupby('Warehouse_block')['Customer_rating'].mean().round(2)
print("CSAT Promedio por Género")
print(Genero)
print("CSAT Promedio por Bloque de Bodega ")
print(Bodega)


#3.2: Clientes Nuevos vs Fidelizados
nuevos = (data['Prior_purchases'] <= 1).sum()
fidelizados = (data['Prior_purchases'] >= 2).sum()
total = len(data)
print("Tasa de Retención")
print(f"Clientes nuevos (0-1 compras):      {nuevos} ({round(nuevos/total*100, 1)}%)")
print(f"Clientes fidelizados (2+ compras):  {fidelizados} ({round(fidelizados/total*100, 1)}%)")

#3.3: Llamadas de Soporte Promedio por Rating
Soporte= data.groupby('Customer_rating')['Customer_care_calls'].mean().round(2)
print("Llamadas de Soporte Promedio por Rating")
print(Soporte)
```


**CONCLUSIONES:**

3.1 -El rating promedio es 2.99 tanto en hombres como en mujeres, sin diferencia por género. Por bloque de bodega tampoco hay variación significativa.

3.2 -No hay clientes nuevos, solo los clientes fidelizados se registra con 100%.

3.3 -Los clientes más satisfechos  hacen más llamadas que los insatisfechos. La diferencia es mínima, lo que sugiere que el volumen de soporte no depende de la satisfacción, sino probablemente del tipo de producto o proceso de compra.
