```python
import pandas as pd
df=pd.read_csv("titanic.csv")
df
df.describe()
#Analisis de Supervivencia
porcentaje_supervivencia = (df['Survived'].sum() / len(df)) * 100

print(f"Porcentaje de supervivencia: {porcentaje_supervivencia}%")
#Analisis de clase
clases = df['Pclass'].value_counts()

print("Cantidad de pasajeros por clase:")
print(clases)

porcentaje_clase = (clases / len(df)) * 100

print("Porcentaje por clase:")
print(porcentaje_clase)
# Analisis por genero
genero = df['Sex'].value_counts()

print("Cantidad de pasajeros por género:")
print(genero)
```

**CONCLUSIONES**
- Hay un porcentaje bajo de supervivencia, indicando la tasa de mortalidad en el vuelo que hubo.
- La mayoria de pasajeros iban en la Pclass 3, luego viene en el 1, y finalmente en el 2. Se puede sugerir que la Pclass 3 es la general.
- Hubo un mayor indice de hombres en el vuelo que mujeres.
